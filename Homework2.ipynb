```python
import numpy as np
import matplotlib.pyplot as plt

def dot_x(x):
    return np.exp(x)
```

```python
x_values = np.linspace(-100,100,1000)
```

```python
y_values = dot_x(x_values)
```

```python
x1_point = 0
x2_point = 0
y1_point = dot_x(x1_point)
y2_point = dot_x(x2_point)
```

```python
plt.plot(x_values, y_values, label=r'$r(x) = e^x$')
plt.scatter([x1_point, x2_point], [y1_point, y2_point], color='red', label='Transcritical Bifurcation')
plt.text(x1_point , y1_point + 0.3, '$x_1$', fontsize=12, color='blue', ha='right')
#plt.text(x2_point, y2_point + 0.3, '$x_2$', fontsize=12, color='blue', ha='right')
plt.xlabel('x')
plt.ylabel('$r(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.75)
plt.legend()
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
