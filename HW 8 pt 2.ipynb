```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
mu = -0.04

def system(t, z):
    x, y = z
    dxdt = -y + mu*x + x*y**2
    dydt = x + mu*y - x**2
    return [dxdt, dydt]
```

```python
t = np.linspace(0, 10, 400)
x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
plt.figure(figsize=(10, 8))

for i in range(len(x)):
    for j in range(len(x[i])):
        sol = solve_ivp(system, [0, 10], [x[i, j], y[i, j]], t_eval=t)
        plt.plot(sol.y[0], sol.y[1], color='blue')  # Trajectory
        plt.plot(sol.y[0, 0], sol.y[1, 0], 'ro')  # Start point
```

```python
Y, X = np.mgrid[-2:2:100j, -2:2:100j]
U = -Y + mu*X + X*Y**2
V = X + mu*Y - X**2
plt.streamplot(X, Y, U, V, color='gray')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase Portrait')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.grid(True)
plt.show()
```
